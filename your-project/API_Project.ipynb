```python
covid_19_countries = response.json()['response'][0:]
covid_19_df = pd.DataFrame(covid_19_countries)


# Displaying the DataFrame
display(covid_19_df)
```

```python
df = covid_19_df.drop(['population','tests'], axis=1, errors='ignore')
display(df)
```

```python
# Extracting the 'active' values from the 'cases' column and creating a new column 'active_cases'
df['active_cases'] = df['cases'].apply(lambda x: x.get('active', None))

# Sorting the DataFrame based on the 'active_cases' column in descending order
df_sorted = df.sort_values(by='active_cases', ascending=False)

# Displaying the sorted DataFrame
print("\nDataFrame sorted by 'active_cases':")
display(df_sorted.head(20))
```

```python
# Grouping by 'continent' and calculating the sum of 'active_cases'
df_continent_total = df.groupby('continent')['active_cases'].sum().reset_index()

# Sorting the new DataFrame by 'active_cases' in descending order
df_continent_total_sorted = df_continent_total.sort_values(by='active_cases', ascending=False)

# Displaying the sorted DataFrame
print("\nDataFrame showing each unique continent with their total of active cases (ordered by active cases - higher to lower):")
display(df_continent_total_sorted)
```

```python
df_zero_active_cases = df[df['cases'].apply(lambda x: x.get('active', None)) == 0]

# Displaying the filtered DataFrame
display(df_zero_active_cases)
```

Web Scraping - TripAdvisor - Restaurants

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('https://www.tripadvisor.com/')
time.sleep(5)
accept = driver.find_element(By.XPATH, '//*[@id="onetrust-accept-btn-handler"]')
accept.click()
time.sleep(3)
restaurants = driver.find_element(By.XPATH, '//*[@id="lithium-root"]/main/div[3]/div[1]/button[4]')
restaurants.click()
search_to_place = driver.find_element(By.XPATH, '//*[@id="lithium-root"]/main/div[4]/div/div/div[2]/div/form/div/div/input')
search_to_place.click()
time.sleep(2)
search_to_place.send_keys(destination)
time.sleep(1)
search_to_place.send_keys(Keys.ENTER)
```

```python
restaurants_page = driver.page_source
time.sleep(5)
soup2 = BeautifulSoup(restaurants_page, 'lxml')
restaurant_names = soup2.find_all('div', attrs={'class': 'result-title'})
addresses_names = soup2.find_all('div', attrs={'class': 'address-text'})
restaurant_rating = soup2.find_all('a', attrs={'class': 'review_count'})

# Extract text from the elements
restaurant_list = [restaurant.text for restaurant in restaurant_names]
addresses_list = [address.text for address in addresses_names]
ratings_list = [rating.text for rating in restaurant_rating]

# Extract text of rating
res_rating = restaurant_rating[0].get_text() if restaurant_rating else 'N/A'

# Add the ratings to the restaurant_addresses dictionary
restaurant_addresses = {}
for i in range(len(restaurant_list)):
    restaurant_addresses[restaurant_list[i]] = {
        'address': addresses_list[i] if i < len(addresses_list) else 'N/A',
        'rating': ratings_list[i] if i < len(ratings_list) else 'N/A',
        'res_rating': res_rating if i == 0 else 'N/A'  # Add res_rating to each entry
    }
```

```python
data = []

for restaurant, info in restaurant_addresses.items():
    # Extract numeric value from 'Rating'
    numeric_rating = int(''.join(filter(str.isdigit, info['rating'])))
    data.append([restaurant, numeric_rating, info['address'], info['rating']])

# Create DataFrame
columns = ["Restaurant Name", "Total Reviews", "Address", "Rating"]
restaurant_df = pd.DataFrame(data, columns=columns)

# Order the DataFrame by the 'Numeric Rating' column in descending order
restaurant_df = restaurant_df.sort_values(by='Total Reviews', ascending=False)

# Drop the 'Rating' column
restaurant_df = restaurant_df.drop(columns=['Rating'])

# Display the DataFrame
display(restaurant_df)
```

```python
# Adding a link to be more friend-user

data = []

for restaurant, info in restaurant_addresses.items():
    # Extract numeric value from 'Rating'
    numeric_rating = int(''.join(filter(str.isdigit, info['rating'])))
    
    # Extract the link using BeautifulSoup
    link_tag = soup2.find('div', text=restaurant).find_parent('div').find('a', class_='review_count')
    restaurant_link = f"https://www.tripadvisor.com{link_tag['href']}" if link_tag else 'N/A'
    
    data.append([restaurant, numeric_rating, info['address'], info['rating'], restaurant_link])

# Create DataFrame
columns = ["Restaurant Name", "Total Reviews", "Address", "Rating", "Restaurant Link"]
restaurant_df = pd.DataFrame(data, columns=columns)

# Order the DataFrame by the 'Numeric Rating' column in descending order
restaurant_df = restaurant_df.sort_values(by='Total Reviews', ascending=False)

# Drop the 'Rating' column
restaurant_df = restaurant_df.drop(columns=['Rating'])

# Make the link clickable in the DataFrame
restaurant_df_styled = restaurant_df.head(5).style.format({'Restaurant Link': '<a href="{0}" target="_blank">{0}</a>'})

# Display only the top 5 rows of the DataFrame
display(HTML(restaurant_df_styled.render()))
```

Web Scraping - Fever - Expeditions

```python
# Set up the webdriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('https://feverup.com/')

# Accept cookie
time.sleep(1)
accept_cookie = driver.find_element(By.XPATH, '//*[@id="cookie-advice"]/button')
accept_cookie.click()

# Click on the city selector
time.sleep(1)
select_city = driver.find_element(By.XPATH, '//*[@id="navbarSupportedContent"]/ul/li')
select_city.click()

# Click on the box to open the city input
time.sleep(1)
box_city = driver.find_element(By.XPATH, '/html/body/astro-island[3]/div[2]/div/div/div/div/div[1]')
box_city.click()

# Input the city in the box
time.sleep(1)
box1_city = driver.find_element(By.XPATH, '/html/body/astro-island[3]/div[2]/div/div/div/div/div[1]/input')
box1_city.send_keys(destination)

# Confirm the city selection
time.sleep(1)
confirm_box_city = driver.find_element(By.XPATH, '/html/body/astro-island[3]/div[2]/div/div/div/div/div[2]/ul/li[1]')
confirm_box_city.click()

# Wait for the city links to load
time.sleep(1)
city_links = driver.find_elements(By.CLASS_NAME, 'svelte-194rbqn')

# Iterate through the links and click on the one that matches the desired city
for city_link in city_links:
    if city_link.text.lower() == destination.lower():
        city_link.click()

# Click on Top 10
time.sleep(1)
top_10 = driver.find_element(By.XPATH, '/html/body/div/main/section[1]/h2/a')
top_10.click()
```

```python
events_page = driver.page_source
time.sleep(5)
soup2 = BeautifulSoup(events_page, 'lxml')  # Corrected variable name
events_names = soup2.find_all('h3', attrs={'class': 'fv-plan-card__title fs-12 fw-bold'})  # Check the class attribute in the actual page source

# Extract text from the elements
event_list = [event.text for event in events_names]

# Create DataFrame
columns = ["Event Name"]
event_df = pd.DataFrame(event_list, columns=columns)

display(event_df)
```

```python
# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(events_page, 'html.parser')

# Find all events
events = soup.find_all('a', {'data-testid': 'fv-plan-card'})

# Extract event names and links
event_list = []
for event in events:
    event_name = event.find('h3', {'class': 'fv-plan-card__title fs-12 fw-bold'}).text
    event_info = event['href']
    event_list.append({'Event Name': event_name, 'Event Info': event_info})

# Create DataFrame
event_df = pd.DataFrame(event_list)

# Select only 'Event Name' and 'Event Link' columns
event_df = event_df[['Event Name', 'Event Info']]

# Convert the 'Event Link' column to clickable links
event_df['Event Info'] = event_df['Event Info'].apply(lambda x: f'<a href="{x}" target="_blank">Link</a>')

# Display the DataFrame with clickable links
display_html(event_df.to_html(escape=False), raw=True)
```
